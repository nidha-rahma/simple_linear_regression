```python
CSV_PATH = "/content/Salary_dataset.csv"
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv(CSV_PATH)
print("First 5 rows:")
display(df.head())
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

```python
# Create a Linear Regression model
model = LinearRegression()

# Train the model (fit the data)
model.fit(X, y)
```

```python
# Get the coefficients (slopes) and intercept
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

```python
# Predictions for existing data
y_pred = model.predict(X)
```

```python
# Predict salary for 5 years of experience
years = [[5]]   # Example new data
predicted_salary = model.predict(years)
print(f"Predicted salary for 5 years experience: {predicted_salary[0]}")
```

```python
plt.scatter(X, y, color="green", label="Actual data")
plt.plot(X, y_pred, color="orange", label="Regression line")
plt.scatter(years, predicted_salary, color="blue", marker="x", label="Prediction (5 yrs)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
